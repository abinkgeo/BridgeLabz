```python
import matplotlib .pyplot as plt
import numpy as np
import pandas as pd
```

```python
#  1 to display a bar chart of the popularity of programming  Languages.

data={"Programming languages":["Java", "Python", "PHP","JavaScript", "C#", "C++"],
      "Popularity":[22.2, 17.6, 8.8, 8, 7.7, 6.7]}

df=pd.DataFrame(data)

plt.bar(df["Programming languages"],df["Popularity"])
plt.title("Popularity of Programing Language")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
#2 to display a horizontal bar chart of the popularity of programming Languages.

plt.barh(df["Programming languages"],df["Popularity"])
plt.title("Popularity of Programing Language")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
#3 to display a bar chart of the popularity of programming Languages. Use uniform color.

plt.bar(df["Programming languages"],df["Popularity"],color="red")
plt.title("Popularity of Programing Language")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
#4 to display a bar chart of the popularity of programming Languages. Use different color for each bar.

colors=['orange','green','blue','yellow','pink','violet']
plt.bar(df["Programming languages"],df["Popularity"],color=colors)
plt.title("Popularity of Programing Language")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
#5 to display a bar chart of the popularity of programmingLanguages. Attach a text label above each bar displaying its popularity (float value).


def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha='center')


colors=['orange','green','blue','yellow','pink','violet']
plt.bar(df["Programming languages"],df["Popularity"],color=colors)
add_labels(df["Programming languages"],df["Popularity"])
plt.title("Popularity of Programing Language")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
# 6 to display a bar chart of the popularity of programming Languages. Make blue border to each bar.


plt.bar(df["Programming languages"],df["Popularity"],edgecolor='blue',linewidth=2.5,color='yellow')
plt.title("Popularity of Programing Language")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
#7 to display a bar chart of the popularity of programming Languages. Specify the position of each bar plot.

position=np.arange(len(df["Programming languages"]))

plt.bar(position,df["Popularity"])
plt.xticks(position,df["Programming languages"])

plt.show()
```

```python
# 8 to display a bar chart of the popularity of programming Languages. Select the width of each bar and their positions.

pos=np.arange(len(df["Programming languages"]))

plt.bar(pos,df["Popularity"], width=0.4)
plt.xticks(pos,df["Programming languages"])
plt.xlabel("Programming Language")
plt.ylabel("Popularity")
plt.show()
```

```python
# 9 to display a bar chart of the popularity of programming Languages. Increase bottom margin.

plt.bar(df["Programming languages"],df["Popularity"])
plt.title("Popularity of Programing Language")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.subplots_adjust(bottom=0.5)
plt.show()
```

```python
#10 to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

data = {
    "Group": ['G1', 'G2', 'G3', 'G4', 'G5'],
    "Men": [22, 30, 35, 35, 26],
    "Women": [25, 32, 30, 35, 29]
}

df=pd.DataFrame(data)

pos=np.arange(len(df["Group"]))
width=0.35

plt.bar(pos - width/2,df["Men"],width,label="Men")
plt.bar(pos + width/2,df["Women"],width,label="Women")

plt.xticks(pos, df["Group"])

plt.xlabel("Groups")
plt.ylabel("Scores")
plt.title("Scores by Group and Gender")
plt.legend()
plt.show()
```

```python
# 11 Write a Python program to create bar plot from a DataFrame.

data = {
    'a': [2, 4, 6, 8, 10],
    'b': [4, 2, 4, 2, 2],
    'c': [8, 3, 7, 6, 4],
    'd': [5, 4, 4, 4, 3],
    'e': [7, 2, 7, 8, 3]
}
df=pd.DataFrame(data)
ax = df.plot(kind='bar')
ax.set_xlabel("Index Values")
ax.set_ylabel("Scores")
ax.set_title("Bar Plot from DataFrame")
plt.show()
```

```python
# 12 Write a Python program to create bar plots with error bars on the same figure.
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_velocity = [0.3314, 0.2278, 0.2836, 0.2645]


x = np.arange(len(mean_velocity))

plt.bar(
    x,
    mean_velocity,
    yerr=std_velocity,
    capsize=5
)

plt.xlabel("Measurements")
plt.ylabel("Mean Velocity")
plt.title("Mean Velocity with Error Bars")


plt.xticks(x, ['A', 'B', 'C', 'D'])


plt.show()

```

```python
# 13 to create bar plots with errorbars on the same figure. Attach a text label above each bar displaying men means (integer value).


plt.bar(x, mean_velocity, yerr=std_velocity, capsize=5)


for i in range(len(mean_velocity)):
    plt.text(
        x[i],
        mean_velocity[i],
        int(mean_velocity[i] * 100),
        ha='center',
        va='bottom'
    )

plt.xlabel("Measurements")
plt.ylabel("Mean Velocity")
plt.title("Mean Velocity with Error Bars")
plt.xticks(x, ['A', 'B', 'C', 'D'])
plt.show()
```
